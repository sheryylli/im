```python
charttimes = []
indexes = []
ids = [] #all the subject
```

```python
#function to place a value in a column
def insertvalue(value, category, desiredindex):
  if category == 'Arterial BP':
    d.loc[desiredindex, 'Arterial BP'] = value
  elif category == 'Arterial BP Mean':
    d.loc[desiredindex, 'Arterial_BP_Mean'] = value
  elif category == 'GCS Total':
    d.loc[desiredindex, 'GCS_Total'] = value
  elif category == 'Heart Rate':
    d.loc[desiredindex, 'Heart_Rate'] = value
  elif category == 'NBP':
    d.loc[desiredindex, 'NBP'] = value
  elif category == 'NBP Mean':
    d.loc[desiredindex,'NBP_Mean'] = value
  elif category == 'Temperature C':
    d.loc[desiredindex, 'Temperature_C'] = value
  elif category == 'Temperature F':
    d.loc[desiredindex,'Temperature F'] = value
  elif category == 'Arterial pH':
    d.loc[desiredindex, 'Arterial pH'] = value
  elif category == 'Hemoglobin':
    d.loc[desiredindex,'Hemoglobin'] = value
  elif category == 'Sodium (135-148)':
    d.loc[desiredindex, 'Sodium (135-148)'] = value
  elif category == 'PAO2':
    d.loc[desiredindex, 'PAO2'] = value
  elif category == 'BUN':
    d.loc[desiredindex, 'BUN'] = value
  elif category == 'Creatinine':
    d.loc[desiredindex, 'Creatinine'] = value
  elif category == 'Potassium':
    d.loc[desiredindex, 'Potassium'] = value
  else:
    #'WBC'
    d.loc[desiredindex,'WBC'] = value
```

```python
for i in range(len(df2)):
  subject = df2.at[i, 'subject_id'] 
  chart = df2.at[i, 'charttime']
  gender = df2.at[i, 'gender']
  weight = df2.at[i, 'weight_first']
  age = df2.at[i, 'icustay_admit_age']
  if subject in ids:
    if chart in charttimes:
      #value = df2.at[i, 'value1']
      #category = df2.at[i, 'label']
      #call function to add value under correct label in new dataframe d
      insertvalue(df2.at[i, 'value1'], df2.at[i, 'label'], indexes[charttimes.index(chart)])
    else:
      #print("yes subject no charttime")
      charttimes.append(chart)
      indexes.append(len(d)-1) #correct index storage for the new array
      #new row with subject and charttime
      #insert the data later with calling new function to add on to the newly created row
      new_row = {'subject_id': subject, 'gender':gender, 'age':age, 'weight':weight, 'charttime':chart}
      #d = d.append(new_row, ignore_index=True)
      elindex = len(d)
      new_row_df = pd.DataFrame(new_row, index={elindex})
      d = pd.concat([d, new_row_df])
      #call the function to add the new value1 to the correct column
      insertvalue(df2.at[i, 'value1'], df2.at[i, 'label'], indexes[charttimes.index(chart)])
  else:
    #print("no subject at all")
    charttimes.clear()
    indexes.clear()
    ids.clear()
    charttimes.append(chart)
    ids.append(subject)
    if i==1:
      indexes.append(0)
    else:
      indexes.append(len(d)-1)
    new_row = {'subject_id': subject, 'gender':gender, 'age':age, 'weight':weight, 'charttime':chart}
    #d = d.append(new_row, ignore_index=True)
    elindex = len(d)
    new_row_df = pd.DataFrame(new_row, index={elindex})
    d = pd.concat([d, new_row_df])
    #call function to add value1 from label into the new d dataframe
    insertvalue(df2.at[i, 'value1'], df2.at[i, 'label'], indexes[charttimes.index(chart)])

print("finished")
```
